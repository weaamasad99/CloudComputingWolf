```python
import json
x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}
print("Python Object: \n\t", x)

for key in x:
    print(key+":\t", x[key])

x_js = json.dumps(x)
print("\nCorresponding JSON Object: \n\t", x_js)

y = json.loads(x_js)
print("\nGoing back to Python: \n")
for key in y:
    print(key+":\t", y[key])
```

```python
import json
from google.colab import drive
import ipywidgets as widgets
from IPython.display import display


# Path to JSON file
file_path = "/content/drive/MyDrive/Students.json"

# Load students
with open(file_path, "r") as f:
    data = json.load(f)

students = data["students"]

# Create list for dropdown
student_names = [s["first_name"] + " " + s["last_name"] for s in students]

dropdown = widgets.Dropdown(
    options=student_names,
    description="Student:",
    style={'description_width': 'initial'}
)

first_name = widgets.Text(description="First Name:")
last_name = widgets.Text(description="Last Name:")
email = widgets.Text(description="Email:")
courses = widgets.Textarea(description="Courses:")
link = widgets.Text(description="Interesting Link:")
favorite_program = widgets.Text(description="Favorite Program:")

update_button = widgets.Button(
    description="Update Program",
    button_style='success'
)

output = widgets.Output()

# Function to display student details
def show_student(change):

    index = student_names.index(dropdown.value)
    student = students[index]

    first_name.value = student["first_name"]
    last_name.value = student["last_name"]
    email.value = student["email"]
    courses.value = ", ".join(student["courses"])
    link.value = student["interesting_link"]
    favorite_program.value = student["favorite_program"]

dropdown.observe(show_student, names='value')

# Update JSON function
def update_program(b):

    index = student_names.index(dropdown.value)
    students[index]["favorite_program"] = favorite_program.value

    with open(file_path, "w") as f:
        json.dump(data, f, indent=4)

    with output:
        output.clear_output()
        print("Program updated successfully!")

update_button.on_click(update_program)

display(dropdown)
display(first_name)
display(last_name)
display(email)
display(courses)
display(link)
display(favorite_program)
display(update_button)
display(output)

# Show first student by default
show_student(None)
```
